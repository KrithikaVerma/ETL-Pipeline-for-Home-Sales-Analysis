```python
# Dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
#storing Indy Homesale csv into Df

homesale_csv = "C:/HOMEWORK/ETL-Project/home_prices_final.csv"
homesale_df = pd.read_csv(homesale_csv)
homesale_df.head()
```

```python
new_homesale_df = homesale_df[[ 'PropStreet', 'PropCity', 'PropState', 'PropZip', 'C6_Sales_price_Assessor']].copy()
new_homesale_df.head(100)
```

```python
# Reference on Empty Check, filled NaN with values 

values = {'PropStreet': "New Address", "PropState" : "IN", "PropZip" : "46011", "C6_Sales_price_Assessor": "0"}
new_homesale_df = new_homesale_df.fillna(value=values)
```

```python
#Storing Indy Weather CSV into df

Indy_Weather_csv = "C:/HOMEWORK/ETL-Project/average_indiana_weather.csv"
IndyWeather_df = pd.read_csv(Indy_Weather_csv)
IndyWeather_df.head()
```

```python
# NaN replaced

indy_Weather_Final = IndyWeather_df.fillna(0)
indy_Weather_Final.head()
```

```python
# Column names were made pretty, removed the space and brackets

indy_Weather_Final.columns = indy_Weather_Final.columns.str.replace(' ', '_')
indy_Weather_Final.columns = indy_Weather_Final.columns.str.replace('(', '')
indy_Weather_Final.columns = indy_Weather_Final.columns.str.replace(')', '')
indy_Weather_Final.head()
```

```python
#Storing Indy Census CSV into DF

IndyCensus_csv = "C:/HOMEWORK/ETL-Project/IN_census_file.csv"
IndyCensus_df = pd.read_csv(IndyCensus_csv)
IndyCensus_df.head()
```

```python
#create connection to DB in postgres, then it works, (view edit data in pgadmin)

rds_connection_string = "postgresql://postgres:Justink59@localhost:5432/HomeSales_DB"


engine = create_engine(rds_connection_string)
```

```python
#check to see the table names

engine.table_names()
```

```python
#Census DataFrame converted into Table 

IndyCensus_df.to_sql(name='Indy_Census_table', con=engine, if_exists = 'append', index=False)
Census_Table_Data = pd.read_sql('SELECT * FROM "Indy_Census_table"', engine)
Census_Table_Data
```

```python
#Homesale DF converted to table
new_homesale_df.to_sql(name='Home_Sale_Table', con=engine, if_exists = 'append', index=False)
Home_Sale_Table_Data = pd.read_sql('SELECT * FROM "Home_Sale_Table"', engine)
Home_Sale_Table_Data
```

```python
#weather df converted to table

indy_Weather_Final.to_sql(name='Indy_Weather_Table', con=engine, if_exists = 'append', index=False)
Indy_Weather_Table_Data = pd.read_sql('SELECT * FROM "Indy_Weather_Table"', engine)
Indy_Weather_Table_Data
```
