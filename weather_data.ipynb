```python
#import dependencies
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.usclimatedata.com/climate/indiana/united-states/3184'
browser.visit(url)
```

```python
cities_list = []

# HTML object
html = browser.html
# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')
    
cities = soup.find_all('div', class_='col-6 col-sm-4 state_name')

# Loop through returned results
for city in cities:
    
    name = city.a.text
    cities_list.append(name)
    
```

The url endpoint was pulled using the following code.

```python
url_list = []

# HTML object
html = browser.html
# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')
    
urls = soup.find_all('div', class_='col-6 col-sm-4 state_name')

# Loop through returned results
for url in urls:
    
    name = url.a
    #print(name['href'])
    url_list.append(name['href'])
    
```

```python
average_monthly = pd.DataFrame()
```

```python
for url in url_list:
    main_url = "https://www.usclimatedata.com" + url
    tables = pd.read_html(main_url)
    df_1 = tables[0]
    df_2 = tables[1]
    df_3 = pd.merge(df_1,df_2, on = "Unnamed: 0",how="inner")
    url_split = url.split("/")
    df_3["City"] = url_split[2].upper()
    average_monthly = average_monthly.append(df_3)
    

  
#/climate/woodburn/indiana/united-states/usin1385
```
