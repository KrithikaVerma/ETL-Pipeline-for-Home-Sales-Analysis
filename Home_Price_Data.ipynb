The city names are converted to upper case to match the entries in the dataset.

```python
city_upper = []
for city in city_list:
    name = city.upper()
    city_upper.append(name)
print(city_upper)   
```

A for loop was used to filter out only the cities that we have weather data for.

```python
final_home = pd.DataFrame()
```

```python
for city in city_upper:
    df = home_filtered.loc[home_filtered["PropCity"]==city]
    final_home = final_home.append(df)
```

```python
final_home.head(25)
```

Columns are renamed to be more descriptive.

```python
home_prices = final_home.rename({"Parcel1_Acreage":"Acreage","PropStreet":"Property Street","PropCity":"Property City",
                  "PropState":"Property State","PropZip":"Property Zip","C6_Sales_price_Assessor":"Sale Price",
                  "Buyer1Street":"Buyer Street","Buyer1City":"Buyer City","Buyer1State":"Buyer State",
                  "Buyer1Zip":"Buyer Zip","Conveyance_Date":"Conveyance Date"})
```

.drop_duplicates was used to remove duplicate data.

```python
home_prices_final = pd.DataFrame.drop_duplicates(home_prices)
home_prices_final
```
